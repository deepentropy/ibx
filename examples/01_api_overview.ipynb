# ib-engine Python API Overview

This notebook demonstrates the ib-engine Python API surface without requiring a live IB connection.

```python
import ibx

# Module contents
public_api = [x for x in dir(ib_engine) if not x.startswith('_')]
print("Public API:", public_api)
```

```python
# Constants - fixed-point scaling factors
print(f"PRICE_SCALE: {ibx.PRICE_SCALE:,}  (1e8 - prices are i64 * 10^8)")
print(f"QTY_SCALE:   {ibx.QTY_SCALE:,}  (1e4 - quantities are i64 * 10^4)")
print()
print("Python binding converts to/from float automatically:")
print(f"  $150.25 in Rust = {int(150.25 * ibx.PRICE_SCALE):,}")
print(f"  100 shares in Rust = {int(100 * ibx.QTY_SCALE):,}")
```

```python
# Available classes
print("IbEngine methods:")
for attr in dir(ibx.IbEngine):
    if not attr.startswith('_'):
        print(f"  .{attr}()")
```

```python
# PyQuote fields (returned by engine.quote())
print("PyQuote fields:")
for attr in dir(ibx.PyQuote):
    if not attr.startswith('_'):
        print(f"  .{attr}")
```

```python
# PyFill fields (returned by engine.fills())
print("PyFill fields:")
for attr in dir(ibx.PyFill):
    if not attr.startswith('_'):
        print(f"  .{attr}")

print()
print("PyOrderUpdate fields:")
for attr in dir(ibx.PyOrderUpdate):
    if not attr.startswith('_'):
        print(f"  .{attr}")

print()
print("PyAccountState fields:")
for attr in dir(ibx.PyAccountState):
    if not attr.startswith('_'):
        print(f"  .{attr}")
```
